# **Visualization:**

```python
albania_data = df[df['Country Name'] == 'Albania']
years = albania_data.columns[4:]
population = albania_data.iloc[0, 4:]

plt.figure(figsize=(12, 6))
plt.bar(years, population)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population of Albania over the Years')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
